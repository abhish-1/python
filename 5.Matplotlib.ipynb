```python
import matplotlib.pyplot as plt
```

```python
X = [1,2,3,4,5,6,7]
Y = [1,2,3,4,5,6,7]
plt.plot(X,Y,label="Directly prop")
plt.title("X directly proportional to Y")
plt.xlabel("X")
plt.ylabel("Y")
X1 = [1,2,3,4,5,6,7]
Y1 = [1,4,9,16,25,36,49]
plt.plot(X1,Y1,label="squarely prop")
plt.legend()
```
