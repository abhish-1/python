```python
### BarChart
#You can type values above each block, shift blocks, make extra space in the right side, ofc change colors.

#draw 2 consecutive graphs next to each other, comparison types.
oscar_revenue=[123,578,2,5,33,56]
non_oscar_revenue=[243,78,23,54,73,96]
years = [2000,2001,2002,2003,2004,2005]

import numpy as np
x = np.arange(len(years)) #->[0,1,2,3,4,5]
width = 0.4
plt.bar(x-width/2, oscar_revenue, width, label = "oscar")
plt.bar(x+width/2, non_oscar_revenue, width, label = "oscar")
# dekho by default width for one block is 1 unit, 0.8 is width of block and 0.1 on both sides for a clenaer look, you can always change it.
# plt.bar me first parameter h centre of the block and third h uska width. Pele 0.8 me ek tha ab 2 h to 0.4 each and centre shift krdena 
# left and right side me that it, make a diag if necesary, will make it cleaner
plt.xlabel("Years")
plt.ylabel("revenue in Million")
plt.title("yearly revenue comparison")
plt.legend()
plt.tight_layout()

plt.xticks(x,years)
```

```python
#scatter plot
age =[12,19,22,33,44]
bp = [100,120,140,150,170]
#A few more changes can  be made on appearences check out their docs
colors=["red" if x>130 else "green" for x in bp]
plt.scatter(age, bp, color = colors)
plt.grid()
plt.xlabel("age")
plt.ylabel("bp")
plt.title("age-BP comparison")

plt.annotate(age[0], xy=(age[0],bp[0]),xytext=(age[0],bp[0]))
```

```python
#PIE chart
expenses = ["salaries", "Rent", "Marketing", "R&D", "misc"]
amount=[100,134,166,234,35]
explode=[0,0,0,0.2,0]
plt.pie(amount, labels=expenses, autopct="%1.1f%%",
        wedgeprops = {
            "edgecolor":"black",
            "linewidth":1,
            "linestyle":"--"
        },
        explode=explode,
        startangle = 180,
        shadow=True)

plt.show()
#https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.Wedge.html
```
