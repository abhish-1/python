```python
URL ="https://www.scrapethissite.com/pages/simple/"
headers = {
    "User-agent": "Mozilla/5.0, headers=headers"
    #sometimes websites expect you to pass the headers (browser do)which can avoid us getting blocked although nowdays itisnt useful
}
res = requests.get(URL)
```

```python
with open("scraped_data/data1.html","w",encoding="utf-8") as f:
    f.write(res.text)
```

```python
import lxml
from bs4 import BeautifulSoup
with open("scraped_data/data1.html","r") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content,"lxml") #lxml is a html parser for BS

#Methods -> find(), find_all(), find_parent(), find_next(), find_previous()
```

```python
all_h3= soup.find_all("h3")
all_countries = []
for h3 in all_h3:
    name = h3.get_text(strip=True)
    population = h3.find_next("div").select("span.country-population")[0].get_text(strip=True)
    all_countries.append([name,population])

print(all_countries)

import pandas as pd
df = pd.DataFrame(all_countries, columns=["Name","Population"])

df.to_csv("scraped_data/cleanedData1.csv",index=False)

```
